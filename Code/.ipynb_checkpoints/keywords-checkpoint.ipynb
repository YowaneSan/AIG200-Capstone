```python
#if needed, runf
#!{sys.executable} -m pip install keybert
import sys
import nltk
import os
from  unidecode  import unidecode
```

```python
from keybert import KeyBERT
import pandas as pd

dataset=[]
paths=[]
keywords=[]
titles=[]
docs=[]
datapaths=[]

#add path to each directory to datapaths
datapaths.append("C:\\Users\\Jeremy Chong\\Desktop\\School\\Capstone\\AIG200-Capstone\\Dataset\\Nervous_System")
datapaths.append("C:\\Users\\Jeremy Chong\\Desktop\\School\\Capstone\\AIG200-Capstone\\Dataset\\Respiratory_System")

for d in datapaths:
    for filename in os.listdir(d):
        if filename.endswith('.txt'): #for each text file
            titles.append(os.path.splitext(filename)[0])
            txt_filename = os.path.splitext(filename)[0] + '.txt'
            txt_path = os.path.join(d, txt_filename)
            
            paths.append(txt_path)
            #print(txt_path)
            with open(txt_path, encoding="utf8") as fo:
                docs.append(unidecode(fo.read())) #append the text to docs
            

#len(docs)
```

```python
kw_model = KeyBERT()
for i in docs:
    keywords.append(kw_model.extract_keywords(i, stop_words='english'))
```

```python
data=[]
for d in range(len(docs)):
    #print(d, len(keywords[d]))
    kw=keywords[d]
    data.append([titles[d], docs[d], kw[0][0], kw[0][1], kw[1][0],kw[1][1],kw[2][0],kw[2][1], kw[3][0],kw[3][1], kw[4][0],kw[4][1]])

#print(data[0][1])
#print(docs[21])
df=pd.DataFrame(data=data, columns=["Title","Text", "Keyword 1","Score1", "Keyword 2", "Score 2", "Keyword 3", "Score 3", "Keyword 4","Score 4","Keyword 5", "Score 5"])
```
